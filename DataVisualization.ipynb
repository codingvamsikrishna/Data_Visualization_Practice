**Apply the colours to the markers**

```python
x = np.array([1,2,3,4,6,8])
y = np.array([3,8,1,5,10,7])
#draw a lineplot insted dot use the "P" and the color is green
plt.plot(x,y,"P--g") #colur = [g,r,b,c,m,y,k,w]
plt.show()
```

**Apply different formatted lines**

```python
#1. dotted line format (:)
#2. solid line format (-)
#3. dashed line format (--)
#4. dash dot lines (-.)
```

```python
#apply dash dit line for above data with increse in marker size--> use the markersize (ms)
plt.plot(x,y,"*-.y",ms= 10)
plt.show()
```

```python
#for the above plot change the marker edge colour
plt.plot(x,y,"*:c",ms=10,mec='b')
plt.show()
```

```python
#change the marker colour to the yellow for above code --> use marker fill color --> mfc
plt.plot(x,y,"*:c",ms=10,mec='b',mfc='#f20cb5')
plt.show()
```

**Grids in plot**

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,4,5,8,9,10]
#in above data x is the entery number for examination and y is the marks obtained in exam
#draw a lineplot with the pentagone and define the xlabel and y label
plt.plot(x,y,'p')
plt.xlabel('entry number')
plt.ylabel('marks in exam')
plt.title("marks vs entry number")
plt.grid()
plt.show()
```

```python
#print the grids on x axis with different line format and different grid line color
plt.plot(x,y,'p')
plt.xlabel('entry number')
plt.ylabel('marks in exam')
plt.title("marks vs entry number")
plt.grid(linestyle="--",linewidth=0.5,color="#f2ad0c",axis="x")
plt.show()
```

```python
#syntax of line plot--> plt.plot(x,y,marker/"p:r",ms,mfc,mec)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#use the default dataset from seaborn libraray
data = sns.load_dataset('tips')
data.head()
```

```python
#draw a lineplot between the tip and total_bill from tips data
plt.plot(data['tip'],data['total_bill'],'o:r')
plt.show()
```

```python
plt.scatter(data['tip'],data['total_bill'])
plt.show()
```

```python
#Draw a histogram on total_bills with 10 bins having and with different colour
plt.hist(data['total_bill'],bins=6,color='green')
plt.show()
#plt.plot(x,y,marker/'o:r',ms,mfc,mec)
#plt.scatter(x,y,size,color)
#plt.hist(x,bins,color)
```

**PieChart**-->also computed based on the frequency

```python
#draw a pie chart on tip and give label as sex
plt.pie(data['tip'],labels=data['sex'])
plt.show()
#aggegarte the total tips given male and female
```

```python
gender_count = data.value_counts('sex')
#print(gender_count)
label = gender_count.index
values = gender_count.values
plt.pie(values,labels=label)
plt.show()
```

```python
data['tip'].values
```

```python
#Draw a piechart on sex and tip to see who has paid maximum percentage tips
tip_by_gender = data.groupby('sex')['tip'].sum()
# print(tip_by_gender)
plt.pie(tip_by_gender.values,labels=tip_by_gender.index,autopct='%1.1f%%')
plt.show()
```

**Scatter plot with sns**

```python
#Note:- for the continous data  distribution we use the scatterplot
sns.scatterplot(data = data, x="total_bill",y="tip")
```

```python
#i want see the total bill and tip distribution in day wise
sns.scatterplot(data = data, x="total_bill",y="tip",hue='day')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#use the default dataset from seaborn libraray
data = sns.load_dataset('tips')
data.head()
```

```python
# draw a histogram based on total_bill and gender
plt.figure(figsize=(5,3))
sns.histplot(data = data,x='total_bill',hue='sex')
```

```python
#load the flights data and draw a hisplot for passenger vs month
data1=sns.load_dataset('flights')
data1.head()
plt.figure(figsize=(5,3))
sns.histplot(data = data1,x="passengers",hue='month',bins=10)
```

**Box Plot**

```python
#what are eleemnts do we have in a box plot?
# 1.Lower limit(5%), 2.Quartile1(25%), 3.Median(50%), 4.Quartile3(75%), 5.Upper limit(95%)
#why the box plot is important?
#Box is mainly used to detect outliers
#Box is used to see the distrubution countiouns data
#Draw a Box plot between the gender and Total_bill
sns.boxplot(data = data,x='sex',y='total_bill')
```

```python
#draw the boxplot for flights data
sns.boxplot(data=data1,x='year',y='passengers')
```

```python
#draw the box plot by using the matplotlib
plt.boxplot(data['total_bill'])
```

**Violin plot**

```python
#draw a violin plot between the total_bill and gender
sns.violinplot(data = data,x='sex',y='total_bill')
```

**Bar Plot**

```python
#Draw a bar plot to visuallize total bill per day
sns.barplot(data = data,x='day',y='total_bill',hue='smoker')
```
